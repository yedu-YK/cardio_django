```python
def bpcc(cl1, cl2):
    bpc = []
    for n,v in zip(cl1,cl2):
        if n < 120 and v < 80:
            val = 1
        elif n>=120 and n<130 and v<80:
            val = 2
        elif n>=130 and n<140 or v>=80 and n<89:
            val =3
        elif n>=140 and n<180 or v>=90 and v<120:
            val = 4
        elif n >180 or v > 120:
            val = 5
        bpc.append(val)
    return bpc
```

```python
def atest(one,two):
    bpc = []
    val = 0
    for n,v in zip(one,two):
        if n < 120 and v < 80:
            val = 1
        elif (n>=120 and n<130) and v<80:
            val = 2
        elif (n>=130 and n<140) or (v>=80 and n<89):
            val =3
        elif (n>=140 and n<180) or (v>=90 and v<120):
            val = 4
        elif n >180 or v > 120:
            val = 5
        bpc.append(val)
    return bpc

ad = atest(df_3.ap_hi, df_3.ap_lo)
print(ad[0])
```
